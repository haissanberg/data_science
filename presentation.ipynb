# Motivation

**Goal:** understand which factors are associated with students’ exam performance.

**Research questions:**
- How are math scores distributed?
- Are there differences by gender?
- Does test preparation help?
- Are subjects related (math, reading, writing)?
- Can we predict math score from other variables?


# Dataset

**Source:** Kaggle — *Students Performance in Exams*

**Rows:** 1000 students
**Variables:**
- Gender
- Race/ethnicity (group A–E)
- Parental level of education
- Lunch (standard vs free/reduced)
- Test preparation course (none vs completed)
- Scores: math, reading, writing (0–100)

**Target for modelling:** math_score

# Data preparation

- Renamed columns for readability (e.g., `math score` → `math_score`)
- Checked missing values (none in this dataset)
- Created an additional feature:
  - `average_score = (math + reading + writing) / 3`

**Why:** easier analysis, cleaner plots, and modelling.
